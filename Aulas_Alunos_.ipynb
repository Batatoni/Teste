Como um rover autônomo pode utilizar Machine Learning para continuar mapeando um terreno desconhecido de forma segura quando seus sensores principais sofrem falhas críticas durante uma missão de bilhoes de dólares?

```python
# Bibliotecas
import numpy as np # Gerar números
from sklearn.tree import DecisionTreeClassifier # Classificação - "cérebro" do rover
from sklearn.metrics import accuracy_score # Acurácia
```

```python
# Parte 1: Simulação do Terreno Marciano
np.random.seed(42) # Configuração de padrão dos dados
terreno_treino = np.random.rand(1000,3) # 0.0 a 1.0
```

```python
terreno_treino
```

```python
# Regra de Perigo: Inclinação + Rugosidade > 1.2, o terreno é perigoso (1), senão é seguro
# Radiação (0) / Inclinação (1) / Rugosidade (2)
status_treino = (terreno_treino[:,1] + terreno_treino[:,2] > 1.2).astype(int)
```

```python
terreno_treino
```

```python
import pandas as pd
df_treino = pd.DataFrame(terreno_treino, columns=['Radiação', 'Inclinação', 'Rugosidade'])
df_treino['Status'] = status_treino
df_treino
```

```python
status_treino
```

```python
# Parte 2: Treinamento
# Criando o modelo
cerebro_rover = DecisionTreeClassifier(max_depth=4)
```

```python
cerebro_rover # Modelo não está treinado
```

```python
# Treinar o modelo
cerebro_rover.fit(terreno_treino, status_treino)
print("Modelo Treinado")
```

```python
cerebro_rover
```

```python
# Parte 3: Simulação de Operação Normal e Falha Crítica
```

```python
terreno_novo = np.random.rand(200,3)
```

```python
# Calcula o status real desse novo terreno para podermos comparar depois
status_real = (terreno_novo[:,1] + terreno_novo[:,2] > 1.2).astype(int)
```

```python
previsao_normal = cerebro_rover.predict(terreno_novo)
acuracia_normal = accuracy_score(status_real, previsao_normal)
print(f"Acurácia no Terreno Normal: {acuracia_normal * 100:.2f}%")
```

```python
terreno_com_falha = terreno_novo.copy()
terreno_com_falha[:,1] = 0.0
```

```python
previsao_falha = cerebro_rover.predict(terreno_com_falha)
acuracia_falha = accuracy_score(status_real, previsao_falha)
print(f"Acurácia no Terreno com Falha Crítica: {acuracia_falha * 100:.2f}%")
```

```python
perda = (acuracia_normal - acuracia_falha) * 100
print(f"Risco financeiro/missão: Houve uma queda de {perda:.1f}% na segurança.")
```

```python
import numpy as np

terreno_treino_emergencia = terreno_treino[:, [0, 2]]
print("Novo conjunto de treinamento criado: terreno_treino_emergencia_sem_inclinacao")
```

```python
from sklearn.tree import DecisionTreeClassifier

# 1. Definir status_treino_emergencia a partir do terreno_treino_emergencia original (3 colunas)
status_treino_emergencia = (terreno_treino_emergencia[:,0] + terreno_treino_emergencia[:,1] > 1.2).astype(int)

# 2. Criar uma nova instância de DecisionTreeClassifier
cerebro_emergencia = DecisionTreeClassifier(max_depth=4)

# 3. Treinar o modelo cerebro_emergencia
cerebro_emergencia.fit(terreno_treino_emergencia, status_treino_emergencia)

print("Modelo cerebro_emergencia treinado com sucesso!")
```

```python


# 1. Calcule a média das leituras da coluna de 'Inclinação' (coluna de índice 1) do array terreno_com_falha.
media_inclinacao = np.mean(terreno_com_falha[:, 1])

# 2. Implemente uma condição if para verificar se essa média é exatamente 0.0.
if media_inclinacao == 0.0:

    terreno_com_falha_sem_inclinacao = np.delete(terreno_com_falha, 1, axis=1)
    previsao_final = cerebro_emergencia.predict(terreno_com_falha_sem_inclinacao)
    print("Condição de falha detectada: Usando o modelo cerebro_emergencia para previsões.")
else:
    previsao_final = cerebro_rover.predict(terreno_com_falha)

    print("Nenhuma falha crítica detectada na inclinação: Usando o modelo cerebro_rover para previsões.")

print("Simulação de diagnóstico e previsão concluída.")
```

```python

# Calcular a acurácia da previsão final
acuracia_final = accuracy_score(status_real, previsao_final)

# Apresentar as acurácias para comparação
print(f"Acurácia no Terreno Normal: {acuracia_normal * 100:.2f}%")
print(f"Acurácia no Terreno com Falha Crítica (Sem ML Adaptativo): {acuracia_falha * 100:.2f}%")
print(f"Acurácia com o Sistema de Decisão de Emergência: {acuracia_final * 100:.2f}%")

# Comparação
if acuracia_final > acuracia_falha:
    print(f"O sistema de emergência melhorou a acurácia em {(acuracia_final - acuracia_falha) * 100:.2f}% em relação à falha crítica.")
else:
    print(f"O sistema de emergência não melhorou a acurácia em relação à falha crítica.")
```
